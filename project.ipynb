```python
import pandas as pd
```

```python
df=pd.read_csv("novagen_dataset.csv")
```

```python
X=df.drop(columns="Target")
y=df["Target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    X,y,
    test_size=0.3,
    random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score,recall_score

lr=LogisticRegression(
    max_iter=1000,
    random_state=42
)

lr.fit(X_train_scaled,y_train)
y_lr_pred=lr.predict(X_test_scaled)


print("classification report:",classification_report(y_test,y_lr_pred))
print("accuracy score",accuracy_score(y_test,y_lr_pred))
print("recall score",recall_score(y_test,y_lr_pred))

```

```python
#KNN 
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(
    n_neighbors=5,
    metric="euclidean"
)
knn.fit(X_train_scaled,y_train)
y_knn_pred=knn.predict(X_test_scaled)

print("classification report:",classification_report(y_test,y_knn_pred))
print("accuracy score",accuracy_score(y_test,y_knn_pred))
print("recall score",recall_score(y_test,y_knn_pred))
```

```python
#random forests classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    random_state=42
)
rf.fit(X_train_scaled,y_train)
y_rf_pred=rf.predict(X_test_scaled)

print("classification report:",classification_report(y_test,y_rf_pred))
print("accuracy score",accuracy_score(y_test,y_rf_pred))
print("recall score",recall_score(y_test,y_rf_pred))
```

```python
#svm
from sklearn.svm import SVC

svc=SVC(
    kernel='rbf',
    max_iter=1000
)
svc.fit(X_train_scaled,y_train)
y_svc_pred=svc.predict(X_test_scaled)

print("classification report:",classification_report(y_test,y_svc_pred))
print("accuracy score",accuracy_score(y_test,y_svc_pred))
print("recall score",recall_score(y_test,y_svc_pred))
```

```python
#ADA boosting
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(
     n_estimators=50,
    learning_rate=1.0,
    random_state=42
)
ad.fit(X_train_scaled,y_train)
y_ad_pred=ad.predict(X_test_scaled)

print("classification report:",classification_report(y_test,y_ad_pred))
print("accuracy score",accuracy_score(y_test,y_ad_pred))
print("recall score",recall_score(y_test,y_ad_pred))
```
